```python
#Recommending items that go together
from itertools import permutations

# Creating the function to find all permutations
def find_pairs(x):
  pairs = pd.DataFrame(list(permutations(x.values, 2)),
                       columns=['a', 'b'])
  return pairs
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

english_stops = stopwords.words('english')

#sampled_beauty_metadata_df['description_phrase'] = sampled_beauty_metadata_df.description.apply(' '.join)
sampled_beauty_metadata_df['description_phrase'] = sampled_beauty_metadata_df['description_phrase'].apply(
    lambda row: [t.lower() for t in nltk.word_tokenize(row) if t.isalpha() and t not in english_stops]
)

wordnet_lemmatizer = WordNetLemmatizer()

sampled_beauty_metadata_df['description_phrase'] = sampled_beauty_metadata_df['description_phrase'].apply(
    lambda row: [wordnet_lemmatizer.lemmatize(t) for t in row]
)
```

```python
# Get the average rating for each user 
avg_ratings = user_ratings_table.mean(axis=1)

# Center each user's ratings around 0
user_ratings_centered =user_ratings_table.sub(avg_ratings, axis=0)

# Fill in all missing values with 0s
user_ratings_centered.fillna(0, inplace=True)

# Print the mean of each column
print(user_ratings_centered.mean(axis=1))
```

```python
user_ratings_centered
```

```python
# Import the required libraries 
from scipy.sparse.linalg import svds
import numpy as np

# Decompose the matrix
U, sigma, Vt = svds(user_ratings_centered)

# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)
print(sigma)
```

```python
# Dot product of U and sigma
U_sigma = np.dot(U, sigma)

# Dot product of result and Vt
U_sigma_Vt = np.dot(U_sigma, Vt)

# Add back on the row means contained in avg_ratings
uncentered_ratings = U_sigma_Vt + avg_ratings.values.reshape(-1, 1)

# Create DataFrame of the results
calc_pred_ratings_df = pd.DataFrame(uncentered_ratings, 
                                    index=user_ratings_table.index,
                                    columns=user_ratings_table.columns
                                   )
# Print both the recalculated matrix and the original 
print(calc_pred_ratings_df)
print(user_ratings_table)
```

```python
sentiment_beauty_df['review_polarity']*= 5
```

```python
user_sentiment_table = sentiment_beauty_df.reset_index().pivot_table(index="reviewerID", columns="asin", values="review_polarity",aggfunc='mean')
```

```python
# Get the average rating for each user 
avg_sentiment = user_sentiment_table.mean(axis=1)

# Center each user's ratings around 0
user_sentiment_centered =user_sentiment_table.sub(avg_sentiment, axis=0)

# Fill in all missing values with 0s
user_sentiment_centered.fillna(0, inplace=True)

# Print the mean of each column
print(user_sentiment_centered.mean(axis=1))
```

```python
# Import the required libraries 
from scipy.sparse.linalg import svds
import numpy as np

# Decompose the matrix
U, sigma, Vt = svds(user_sentiment_centered)

# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)
print(sigma)
```

```python
 # Dot product of U and sigma
U_sigma = np.dot(U, sigma)

# Dot product of result and Vt
U_sigma_Vt = np.dot(U_sigma, Vt)

# Add back on the row means contained in avg_ratings
uncentered_sentiment = U_sigma_Vt + avg_sentiment.values.reshape(-1, 1)

# Create DataFrame of the results
calc_pred_sentiment_df = pd.DataFrame(uncentered_sentiment, 
                                    index=user_sentiment_table.index,
                                    columns=user_sentiment_table.columns
                                   )
# Print both the recalculated matrix and the original 
print(calc_pred_sentiment_df)
print(user_sentiment_table)
```
